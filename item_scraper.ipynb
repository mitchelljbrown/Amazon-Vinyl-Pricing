```python
# importing libraries
from bs4 import BeautifulSoup
import requests
import time
import datetime
import pandas as pd

import smtplib  #for sending emails
```

#### Scraping Individual Item Data From Amazon.ca

```python
# connect to amazon.ca

URL = 'https://www.amazon.ca/Discovery-Vinyl-Daft-Punk/dp/B099ZRXZ6Q/ref=sr_1_1?m=A3DWYIK6Y9EEQB&pf_rd_i=1292212011&pf_rd_m=A1IM4EOPHS76S7&pf_rd_p=e647b509-1d9d-4abc-80c7-7c085f4e9577&pf_rd_r=2VST4DD7Q3P0P0GDV05K&pf_rd_s=merchandised-search-3&pf_rd_t=101&qid=1646254795&refinements=p_6%3AA3DWYIK6Y9EEQB%2Cp_69%3A90x-0y&s=music&sr=1-1'

headers = {"User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/98.0.4758.102 Safari/537.36", "Accept-Encoding":"gzip, deflate", "Accept":"text/html,application/xhtml+xml,application/xml;q=0.9,*/*;q=0.8", "DNT":"1","Connection":"close", "Upgrade-Insecure-Requests":"1"}

page = requests.get(URL, headers=headers)

soup1 = BeautifulSoup(page.content, "html.parser")

soup2 = BeautifulSoup(soup1.prettify(), "html.parser")

title = soup2.find(id = 'productTitle').get_text().strip()
artist = soup2.find('span', class_="author notFaded").find('a', class_='a-link-normal').get_text().strip()
rating = soup2.find('i', class_='a-icon a-icon-star a-star-5').get_text().strip().split()[0]
num_ratings = soup2.find(id = 'acrCustomerReviewText').get_text().strip().split()[0]
price = soup2.find('span', class_='a-offscreen').get_text().strip()[1:]
info = soup2.find('ul', class_='a-unordered-list a-nostyle a-vertical a-spacing-none detail-bullet-list').get_text().strip().split()
release_year = info[26]
lp_num = int(info[-1])
if lp_num == 2:
    doublelp = True
else:
    doublelp = False
today = datetime.date.today()

print(title)
print(artist)
print(rating)
print(num_ratings)
print(price)
print(release_year)
print(lp_num)
print(today)

# ideas for comment scraping
    # poor quality vinyl
    # average word count compared to rating
```

#### Function to scrape item data

```python
def update_price():
    URL = 'https://www.amazon.ca/Discovery-Vinyl-Daft-Punk/dp/B099ZRXZ6Q/ref=sr_1_1?m=A3DWYIK6Y9EEQB&pf_rd_i=1292212011&pf_rd_m=A1IM4EOPHS76S7&pf_rd_p=e647b509-1d9d-4abc-80c7-7c085f4e9577&pf_rd_r=2VST4DD7Q3P0P0GDV05K&pf_rd_s=merchandised-search-3&pf_rd_t=101&qid=1646254795&refinements=p_6%3AA3DWYIK6Y9EEQB%2Cp_69%3A90x-0y&s=music&sr=1-1'

    headers = {"User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/98.0.4758.102 Safari/537.36", "Accept-Encoding":"gzip, deflate", "Accept":"text/html,application/xhtml+xml,application/xml;q=0.9,*/*;q=0.8", "DNT":"1","Connection":"close", "Upgrade-Insecure-Requests":"1"}

    page = requests.get(URL, headers=headers)

    soup1 = BeautifulSoup(page.content, "html.parser")

    soup2 = BeautifulSoup(soup1.prettify(), "html.parser")

    title = soup2.find(id = 'productTitle').get_text().strip()
    artist = soup2.find('span', class_="author notFaded").find('a', class_='a-link-normal').get_text().strip()
    rating = soup2.find('i', class_='a-icon a-icon-star a-star-5').get_text().strip().split()[0]
    num_ratings = soup2.find(id = 'acrCustomerReviewText').get_text().strip().split()[0]
    price = soup2.find('span', class_='a-offscreen').get_text().strip()[1:]
    info = soup2.find('ul', class_='a-unordered-list a-nostyle a-vertical a-spacing-none detail-bullet-list').get_text().strip().split()
    release_year = info[26]
    lp_num = int(info[-1])
    if lp_num == 2:
        doublelp = True
    else:
        doublelp = False
    today = datetime.date.today()

    header = ['Artist', 'Album', "Rating", "Number of Ratings", "Price", "Release Date", "DoubleLP", "Date"]
    data = [artist, title, rating, num_ratings, price, release_year, lp_num, today]

    with open('webscraping_amazon_vinyl.csv', 'a+', newline='', encoding='UTF8') as f:
        writer = csv.writer(f)
        writer.writerow(data)

    # if (float(price) < 30):
    #     send_email()
```
